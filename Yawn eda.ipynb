```python
import os
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import seaborn as sns
from glob import glob
```

```python
data_dir = r'C:\Users\nandi\OneDrive\Desktop\Minor Project\yawn dataset'
```

```python
train_yawn = glob(os.path.join(data_dir, 'yawn/*.jpg'))  # Assuming your yawning images are in this path
train_no_yawn = glob(os.path.join(data_dir, 'no yawn/*.jpg'))  # Assuming your non-yawning images are in this path
train_new_yawn = glob(os.path.join(data_dir, 'new_yawn/*.jpg'))  # Assuming your yawning images are in this path
train_new_no_yawn = glob(os.path.join(data_dir, 'new_no_yawn/*.jpg'))  # Assuming your yawning images are in this path
```

```python
image_paths = train_yawn + train_no_yawn
```

```python
print(f"Total number of images: {len(image_paths)}")
print(f"Number of imagesfor yawn: {len(train_yawn)}")
print(f"Number of images for no yawn: {len(train_no_yawn)}")
```

```python
def display_images(train_yawn, n_images=5):
    # Check if there are enough images to display
    if len(train_yawn) < n_images:
        print(f"Not enough images to display. Found only {len(train_yawn)} images.")
        n_images = len(image_paths)
    
    # Display the images
    plt.figure(figsize=(15, 10))
    for i in range(n_images):
        img = Image.open(train_yawn[i])
        plt.subplot(1, n_images, i+1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()
```

```python
display_images(train_yawn, n_images=5)
```

```python
def display_images(train_no_yawn, n_images=5):
    # Check if there are enough images to display
    if len(train_no_yawn) < n_images:
        print(f"Not enough images to display. Found only {len(train_no_yawn)} images.")
        n_images = len(image_paths)
    
    # Display the images
    plt.figure(figsize=(15, 10))
    for i in range(n_images):
        img = Image.open(train_no_yawn[i])
        plt.subplot(1, n_images, i+1)
        plt.imshow(img)
        plt.axis('off')
    plt.show()
```

```python
display_images(train_no_yawn, n_images=5)
```

```python
import os
from PIL import Image

# Define the directory containing your images
yawn_dir = r'C:\Users\nandi\OneDrive\Desktop\Minor Project\yawn dataset\yawn'  # Update this with the actual path to your yawn images
no_yawn_dir = r'C:\Users\nandi\OneDrive\Desktop\Minor Project\yawn dataset\no yawn'  # Update this with the actual path to your no-yawn images
output_dir_yawn = r'C:\Users\nandi\OneDrive\Desktop\Minor Project\yawn dataset\new_yawn'  # Output directory for resized yawning images
output_dir_no_yawn = r'C:\Users\nandi\OneDrive\Desktop\Minor Project\yawn dataset\new_no_yawn'  # Output directory for resized non-yawning images

# Create output directories if they don't exist
os.makedirs(output_dir_yawn, exist_ok=True)
os.makedirs(output_dir_no_yawn, exist_ok=True)

# Define target size
target_size = (80, 80)  # Width, Height

# Function to resize images
def resize_images(input_dir, output_dir):
    for img_name in os.listdir(input_dir):
        img_path = os.path.join(input_dir, img_name)
        try:
            img = Image.open(img_path)
            img = img.resize(target_size)  # Resize image
            img.save(os.path.join(output_dir, img_name))  # Save resized image
        except Exception as e:
            print(f"Error processing {img_name}: {e}")

# Resize yawning images
resize_images(yawn_dir, output_dir_yawn)

# Resize non-yawning images
resize_images(no_yawn_dir, output_dir_no_yawn)

print("Images have been resized and saved to the output directories.")
```

```python
image_shapes = []
for img_path in train_new_yawn[:100]:  # Checking shapes of first 100 images
    img = Image.open(img_path)
    image_shapes.append(img.size)

image_shapes = np.array(image_shapes)

# Separate the widths and heights
widths = image_shapes[:, 0]
heights = image_shapes[:, 1]

plt.figure(figsize=(10, 5))
sns.violinplot(data=[widths, heights], palette='muted')
plt.xticks(ticks=[0, 1], labels=['Width', 'Height'])
plt.title('Violin Plot of Image Sizes')
plt.ylabel('Pixels')
plt.show()
```
